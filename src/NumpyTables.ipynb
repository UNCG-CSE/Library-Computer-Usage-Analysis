UNCG's Spring Semester 2017 started on 1/9/17 and ended on 5/12/17. According to below, this is 2953 hours.

```python
pd.date_range(start='01-09-2017',end='5-12-2017',freq='h')
```

```python
computerAccessTimes = pd.date_range(start='01-09-2017',end='5-12-2017',freq='h')
```

```python
type(computerAccessTimes)
```

The number of machines in the library is dependent on a table of values that correlate primary key to machine name. There are 312 entries.

```python
computerInfo = pd.read_csv(r'../data/libraryMachines.csv')
```

```python
computerInfo.info()
```

Since each one of these cells will contain the number of minutes a machine is used per hour, a 16-bit integer should be suitable (n <= 60).

```python
computerDataGrid = pd.DataFrame(index=computerAccessTimes,columns=computerInfo.machineName,dtype='int16')
```

```python
computerDataGrid
```

A pandas table with 2953 rows and 312 columns and int16 cells takes 7.1 MB.
